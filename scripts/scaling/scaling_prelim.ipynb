```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter1d
from functools import partial
```

```python
smooth = partial(uniform_filter1d, mode="nearest", size=10)
```

```python
loss300 = np.genfromtxt("loss_300M.txt", delimiter=",")
loss700 = np.genfromtxt("loss_700M.txt", delimiter=",")
loss010 = np.genfromtxt("loss_010M.txt", delimiter=",")
loss003 = np.genfromtxt("loss_003M.txt", delimiter=",")

f, axs = plt.subplots(1, 2, ticksin=True, sharey=True, constrained_layout=True)
axs[0].plot(np.log10(loss003[:, 0]*163800*6*3400000),   np.log10(smooth(loss003[:, 1])), ":", label="003M")
axs[0].plot(np.log10(loss010[:, 0]*163800*6*13300000),  np.log10(smooth(loss010[:, 1])), "--", label="013M")
axs[0].plot(np.log10(loss300[:, 0]*163800*6*335000000), np.log10(smooth(loss300[:, 1])), "-.", label="330M")
axs[0].plot(np.log10(loss700[:, 0]*163800*6*740000000), np.log10(smooth(loss700[:, 1])), label="700M")
axs[0].set_ylabel("Huber loss (log10)")
axs[0].set_xlabel("FLOPS (log10)")
axs[0].set_ylim(-3.0, -2.9)
axs[0].set_xlim(16.5, 19.8)

axs[1].plot(loss003[:, 0]*163800/1e9, np.log10(smooth(loss003[:, 1])), ":", label="003M")
axs[1].plot(loss010[:, 0]*163800/1e9, np.log10(smooth(loss010[:, 1])), "--",label="013M")
axs[1].plot(loss300[:, 0]*163800/1e9, np.log10(smooth(loss300[:, 1])), "-.", label="330M")
axs[1].plot(loss700[:, 0]*163800/1e9, np.log10(smooth(loss700[:, 1])), label="700M")
axs[1].set_xlabel("Tokens seen (B)")
axs[1].legend(framealpha=0)

f.suptitle("Scaling plots for four preliminary runs \non 1M unique galaxy images (~1B tokens)")
plt.show()
```
